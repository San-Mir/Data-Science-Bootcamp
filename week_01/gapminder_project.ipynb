```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
fertility1 = pd.read_csv('data/gapminder_total_fertility.csv', index_col=0)
life_expec1 = pd.read_excel('data/gapminder_lifeexpectancy.xlsx', index_col=0)
population1  = pd.read_excel('data/gapminder_population.xlsx', index_col=0)
```

```python
print(fertility1.shape)
print(life_expec1.shape)
print(population1.shape)
```

```python
for i in range(2):
    population1.dropna(axis=i, how="all", inplace=True)
    fertility1.dropna(axis=i, how="all", inplace=True)
    life_expec1.dropna(axis=i, how="all", inplace=True)
```

```python
print(fertility1.shape)
print(life_expec1.shape)
print(population1.shape)
```

```python
fertility1.columns = fertility1.columns.astype(int)
population1.columns = population1.columns.astype(int)
```

```python
print(fertility1.columns)
print(life_expec1.columns)
print(population1.columns)
```

```python
print(fertility1.index)
print(life_expec.index)
print(population1.index)
```

```python
fertility1.index.name = 'country'
life_expec1.index.name = 'country'
population1.index.name = 'country'
```

```python
population1 = population1.reset_index()
fertility1 = fertility1.reset_index()
life_expec1 = life_expec1.reset_index()
```

```python
fertility1 = fertility1.melt(id_vars='country', var_name='year', value_name='fertility1')
population1 = population1.melt(id_vars='country', var_name='year', value_name='population1')
life_expec1 = life_expec1.melt(id_vars='country', var_name='year', value_name='life_expec1')   
```

```python
fertility1.head()
```

```python
population1.head()
```

```python
life_expec1.head()
```

```python
df = pd.merge(fertility1, population1, on=['country', 'year'])
df = pd.merge(df, life_expec1, on=['country', 'year'])
# df = pd.merge(df, continents, on='country', how='outer')
df.iloc[100:105]
```

```python
df.shape
```

```python
df.dropna(inplace=True)
df
```

```python
df_sub = df.loc[df['year'] == 2000]
sns.scatterplot(x='life_expec1', y='fertility1', data=df_sub, alpha=0.6)
```

```python
COLORS = {
    'Europe': 'blue',
    'North America': 'orange',
    'South America': 'yellow',
    'Asia': 'green',
    'Africa': 'red',
    'Australia and Oceania': 'teal'
}
```

```python
# create frames
# written iteratively
for year in range(1960, 2016): # add ,10 for bigger steps -> faster when debugging
    print(year, end=', ')  # for debugging
    subdf = df[df['year'] == year]
    colors= subdf['fertility1'].replace(COLORS).values
    
    plt.figure(figsize=(12, 8)) # <-- reset figure, crashes without this line
    plt.scatter(x=subdf['life_expec1'], y=subdf['fertility1'], 
                c=colors,
                s=subdf['population1'] // 100_000,
                alpha=0.5)
    plt.axis([0, 85, 0, 9.5])
    plt.title(f"{year}")
    plt.xlabel('life expectancy [years]')
    plt.ylabel('fertility [children per women]')
    plt.savefig(r'C:\Users\hassa\OneDrive\Desktop\spiced_excercises\scikit-cilantro-student-code\week_01\data\fig_plot\frame_{year}.png',dpi=90)
    plt.close() 
```
