```python
cur_dir=os.getcwd()
files=[os.path.join(cur_dir,x) for x in os.listdir(cur_dir) if x=="datapoints.csv"]
file_=files[0]
df=pd.read_csv(file_)
df.rename(columns={"y":"ytrue"},inplace="True")

#----------------TASK 3-----------------------------------------
df["y"]=10*df.x+0  # implemenation of the equation for line

#----------Task 4--------------
mse_=np.square(df.y-df.ytrue).sum()/df.shape[0] # error calculated without using a library function

# --------------TASK 5----------------------
list_mse=[]
list_a=list(np.linspace(10,0,100))
list_mse=[np.square(a*df.x+0-df.ytrue).sum()/df.shape[0] for a in list_a] #
(index,min_mse)=(list_mse.index(min(list_mse)),min(list_mse))
result_dict={"Index":index,"val A":list_a[index],"MSE min":min_mse}
plt.plot(list_a,list_mse)
#plt.show()

# --------------TASK 6----------------------
list_b=list(np.linspace(0,1,100))
df1=pd.DataFrame(columns=["A","B","MSE"])
for b in list_b:
    list_mse=[np.square(a*df.x+b-df.ytrue).sum()/df.shape[0] for a in list_a]
    df1=df1.append({'A': list_a[list_mse.index(min(list_mse))],'B': b,"MSE":min(list_mse)}, ignore_index=True)
global_min_index=df1.MSE.idxmin()
global_min=df1.iloc[global_min_index]
print("Min A Val: ",global_min.A,"Min B Val: ",global_min.B,"Min MSE Val: ",global_min.MSE) # final results


#----------PLOTTING------------------
df.plot.line(x="x",y="y",label="true data",legend="True")
#plt.show()
#filepath=

# --------------TASK 7------------------------

# In Task 5 and 6, one can always reduce the size of values a and b gradually, once the code starts to appraoch the desired outcome. 
```
